apply function

```python
#Create the Result column.
clean_std.columns
```

```python
def result(marks):
  if marks >= 250:
    return 'Pass'
  else:
    return 'Fail'
```

```python
clean_std['Result'] = clean_std['Total_Marks'].apply(result)
```

```python
clean_std
```

```python
#Create the Grade column.
def grade(marks):

    if marks >= 450:
        return 'A'

    elif marks >= 400:
        return 'B'

    elif marks >= 350:
        return 'C'

    elif marks >= 300:
        return 'D'

    elif marks >= 250:
        return 'E'

    else:
        return 'F'
```

```python
clean_std['Grade'] = clean_std['Total_Marks'].apply(grade)
```

```python
clean_std['Grade']
```

```python
clean_std.columns
```

```python
def Performance(percentange):
  if percentange >= 90:
    return 'Excellent'
  elif percentange >= 75 and percentange < 90:
    return 'Good'
  elif percentange >= 60 and percentange < 75:
    return 'Average'
  else:
    return 'Poor'
```

```python
clean_std['Performance'] = clean_std['Percentage'].apply(Performance)
```

```python
clean_std[['Student_Name','Average_Marks','Grade','Performance']]
```

```python
#How many students got
#A
#B
#C
#D
#F
clean_std['Grade'].value_counts()
```

lamda function

```python
#scholarship
clean_std['Scholarship'] = clean_std['Percentage'].apply( lambda x: 'Eligible' if x > 85 else 'Not Eligible')
```

```python
#Attendance_Status
clean_std['Attendance_Status'] = clean_std['Attendance'].apply(lambda x: 'Good' if x > 75 else 'Low')
```

```python
#Study_Level
clean_std['Study_Level'] = clean_std['Study_Hours'].apply(lambda x : 'High' if x >= 6 else 'Low')
```

```python
#Performance_Level
#≥90 → Excellent
#≥75 → Good
#≥60 → Average
#Otherwise → Poor
clean_std['Performance_Level'] = clean_std['Percentage'].apply(lambda x : 'Excellent' if x >= 90 else 'Good' if x >=75 else 'Average' if x >= 60 else 'Poor')
```
